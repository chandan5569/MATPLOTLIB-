```python
import numpy as np
import matplotlib.pyplot as pnp

#Here we are setting the axis in terms of X and Y by creating their respective values in Numpy
x=np.arange(0,11)
y=x**2

#Plotting by the Basic function (Plot)


pnp.plot(x,y)
pnp.xlabel('X axis')
pnp.ylabel('Y axis')
pnp.title('Chandan')
```

```python
#use of the Subplot in Functional method

#Syntax of subplot is (nrows= ,ncols= ,plotnumber= )

pnp.subplot(1,2,1)
pnp.plot(x,y)
pnp.xlabel('X')
pnp.ylabel('Y')
pnp.title('plotnumber1')

pnp.subplot(1,2,2)
pnp.plot(y,x)
pnp.xlabel('C')
pnp.ylabel('H')
pnp.title('plotnumber2')

pnp.tight_layout()                  #For proper spcaing between the graphs
```

```python
#OOPS method (object oreinted programming method)

f=pnp.figure()
axes=f.add_axes([0.1,0.1,1,1])  #An empty canvas is created ,Syntax is left, bottam, width, height in a list.
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('chandan')
```

```python
#use of Subplots in OOP method

#Here we use the TUPLE UNPACKING CONCEPT

fig,axes=pnp.subplots(1,2)

axes                             #Here axes is an array
```

```python
fig,axes=pnp.subplots(1,2)
axes[0].plot(x,y)                     #as the axes is the array so we can target it by index number
axes[1].plot(y,x)
```

```python
fig,axes=pnp.subplots(1,2)
axes[0].plot(x,y)                     #as the axes is the array so we can target it by index number
axes[1].plot(y,x)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('plot1')
axes[1].set_xlabel('w')
axes[1].set_ylabel('d')
axes[1].set_title('plot2')
pnp.tight_layout()
```

```python
#use of the legend feild

f=pnp.figure()

axes=f.add_axes([0.1,0.1,1,1])
axes.plot(x,y,label='firstplot')
axes.plot(y,x,label='secondplot')
axes.legend()                           #Position of the legend is by default in the upper right corner
```
