```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
%matplotlib inline
f=plt.figure()
a=f.add_axes([0,0,1,1])
a.plot(x,y)
a.set_xlabel('X')
a.set_ylabel('Y')
a.set_title('TITLE')
```

```python
f=plt.figure()
a1=f.add_axes([0,0,1,1])    #main axes
a2=f.add_axes([0.1,0.5,0.3,0.4])  #inset axes
a1.plot(x,y,color='red')
a2.plot(x,y,color='red')
a1.set_xlabel('x')
a1.set_ylabel('y')
a1.set_title('title')
a2.set_xlabel('x')
a2.set_ylabel('y')
a2.set_title('Title')
```

```python
f=plt.figure()
a1=f.add_axes([0,0,1,1])
a2=f.add_axes([0.1,0.5,0.4,0.4])

import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
a1.plot(x,z)
a2.plot(x,y)
a2.set_xlim([20,22])
a2.set_ylim([30,50])
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
axes[0].plot(x,y,lw=2,ls='--',color='blue')
axes[1].plot(x,z,color='red')
plt.tight_layout()
```
